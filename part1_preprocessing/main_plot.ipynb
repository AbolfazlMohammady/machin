```python
import matplotlib.pyplot as plt
```

```python
# نمودار خطی
x = [0,1,2,2,5,8,4,6,4,6,9]
y = [8,9,4,6,7,8,4,6,4,0,5]
plt.plot(x,y, marker='*',color='red')
plt.title('Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# نمودار پایچارت یا میله ای
x = ['red', 'blue', 'green', 'black']
y = [3, 14, 5, 23]
plt.bar(x,y, color='orange')
plt.title('Example')
plt.xlabel('color')
plt.ylabel('count')
plt.show()
```

```python
# نمودار دایره ای
x = ['red', 'blue', 'green', 'black','sara']
y = [3, 14, 5, 23,14]
plt.pie(y, labels=x ,autopct='%1.1f%%',colors=['red', 'blue', 'green', 'brown','orange'])
plt.show()
```

```python
import numpy as np
x = np.random.rand(1000,1)
plt.hist(x, bins=10)
plt.show()
```

```python
x = np.random.rand(64,64) # یک ماتریک 64 * 64 میسازیم

plt.imshow(x, cmap='hot')   #برای نشون دادن عکس یا مارتریکس ار این استفاده میکنیم
plt.colorbar()
plt.show()
```
