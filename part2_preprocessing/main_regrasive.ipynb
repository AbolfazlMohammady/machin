```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
#data
# Y = 2 * X - 1
x = np.array([1,2,3,4]).reshape(-1,1)
y = np.array([1,3,5,7])
```

```python
x.shape
```

```python
x
```

```python
#plot

plt.plot(x, y, '*')
plt.show()
```

```python
# model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
# Y = m * x +- c
m = model.coef_[0]
c = model.intercept_

print(m, c)
```

```python
plt.plot(x, y, '*')
plt.plot(x, model.predict(x))
plt.show()
```

```python
result = model.predict([[5]])
result
```

```python
yers_of_experience = np.array([0,1,2,3,4,6,8,10]).reshape(-1,1)
salaries= np.array([30000,40000,55000,60000,70000,80000,85000,87000])
```

```python
plt.plot(yers_of_experience, salaries, '*')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor # رگرسیون چند  خطی 

model = DecisionTreeRegressor(max_depth=4) # max_depth عمق 
model.fit(yers_of_experience,salaries)

y_predict = model.predict(yers_of_experience)

plt.plot(yers_of_experience, salaries, '*')
plt.plot(yers_of_experience,y_predict)
plt.show()
```
